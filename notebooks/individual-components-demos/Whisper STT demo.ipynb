```python
import gradio as gr
import whisper
#from time import sleep

# 2. Cargar el modelo Whisper Small (multilingüe) – descarga automática la primera vez
model = whisper.load_model("small")  # modelo "small" ~244 MB, balance precisión/velocidad :contentReference[oaicite:0]{index=0}

# 3. Mapeo de etiquetas a códigos ISO-639-1 que usa Whisper
LANG_CODE = {
    "Español": "es",
    "English": "en"
}

def transcribir(idioma: str, audio: str) -> str:
    """
    - idioma: "Español" o "English"
    - audio: ruta al archivo WAV grabado por Gradio
    Devuelve la transcripción de Whisper Small, forzando el idioma elegido.
    """
    code = LANG_CODE[idioma]
    # El método transcribe admite el parámetro `language` para forzar el idioma :contentReference[oaicite:1]{index=1}
    #sleep(5)
    result = model.transcribe(audio, language=code)
    return result["text"].strip()

# 4. Definición de la interfaz Gradio
iface = gr.Interface(
    fn=transcribir,
    inputs=[
        gr.Dropdown(choices=list(LANG_CODE.keys()),
                    value="Español",
                    label="Idioma"),
        gr.Audio(sources="microphone",
                 type="filepath",
                 label="Grabación")
    ],
    outputs=gr.Textbox(label="Transcripción"),
    title="Transcriptor de Voz con Whisper Small",
    description=(
        "Selecciona el idioma de la grabación, habla por el micrófono "
        "y obtén la transcripción generada por Whisper Small."
    ),
    allow_flagging="never"
)

if __name__ == "__main__":
    iface.launch(debug=True)
```
